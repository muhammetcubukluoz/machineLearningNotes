```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv("6-bank_customers.csv")
```

```python
X = data.drop("subscribed", axis=1)
y = data["subscribed"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()

logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
#score = accuracy_score(y_pred, y_test) #can also be used this way on small data sets
score = accuracy_score(y_test, y_pred)

print("Score: ", score)

print(classification_report(y_pred, y_test))

print("Confusion Matrix: \n",confusion_matrix(y_pred, y_test))
```

```python
# hyperparameter tuning
```

```python
model = LogisticRegression()
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

penalty = ["l1","l2","elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga", "newton-cholesky"]
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

```python
# grid search cv
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold() # n_splits=d default value
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
# https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-api-overview !"scoring"

# n_jobs=-1, Use as much space as you have on your CPU

grid = GridSearchCV(estimator=model, param_grid=params, cv=cv, scoring="accuracy", n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)

print("Score: ", score)

print(classification_report(y_pred, y_test))

print("Confusion Matrix: \n",confusion_matrix(y_pred, y_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html

randomcv = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, n_iter=10, scoring="accuracy")
```
